```python
import pandas as pd
import requests

# Dataset by Jon Bruner https://github.com/jonbruner/twitter-analysis
users = pd.read_csv('users.csv', encoding='latin-1', index_col='account_id')

users = users.astype({ 'url': 'string' })

users = users.dropna()

len(users)

urls = users['url'].tolist()

for url in urls:
    try:
        resp = requests.get(url)
        print(url, resp.status_code, resp.url)
    except:
        print(f'Something went wrong with: {url}')
```
